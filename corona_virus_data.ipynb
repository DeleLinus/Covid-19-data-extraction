```python
import requests
from bs4 import BeautifulSoup

# headers="{'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/79.0.3945.130 Safari/537.36'}"
codes_object=requests.get("https://www.worldometers.info/coronavirus/#countries")#, headers=headers)
c=codes_object.content
c

soup=BeautifulSoup(c,"html.parser")

# datum=soup.find_all("table",{"class":"table"})
# datum0=datum[0].find_all("tr")
# datum0

datum=soup.find_all("tr")
#len(datum)
```

```python
d=datum[0].find_all("th")
#d
for item in d:
    print(item.text)
for item in datum:
    d=item.find_all("td")
    for j in d:
        if j.text!="":
            print(j.text.strip("\n"))
        else:
            print(None)
    print("______________________________")
```

```python
data=[]
dic={}
d=datum[0].find_all("th")
header=[]
#list of headers for easy mapping with the data
for item in d:
    header.append(item.text)
#mapping
for item in datum:
    r=item.find_all("td")
    for head,j in zip(header,r):
        if j.text!="":
            dic[head]=j.text
        else:
            dic[head]=None
#making the list of the dictionary for easy conversion to dataframe
    data.append(dic)
print(data)
```

```python
import pandas as pd
df=pd.DataFrame(data)
df
```
